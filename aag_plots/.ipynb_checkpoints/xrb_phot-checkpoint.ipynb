```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14

# Set tick size
plt.rcParams['xtick.major.size'] = 8
plt.rcParams['ytick.major.size'] = 8
plt.rcParams['xtick.minor.size'] = 4
plt.rcParams['ytick.minor.size'] = 4

plt.rcParams['axes.labelsize'] = 14
```

```python
df = pd.read_csv('data/MAXI1803-293_phot.csv')
```

```python
df2015 = df.query('mjd > 57000')
df2013 = df.query('mjd < 57000')
```

```python
def plot_phot(data, x, y, yerr, band, ax):
    cmap = {'i':'pink', 'g': 'green'}
    for cc in data[band].unique():
        sub = data.query(f'{band} == "{cc}"')
        ax.errorbar(sub[x], sub[y], sub[yerr], color=cmap[cc], label=cc, marker='o', linestyle='none')
```

### 2013 DECam data

```python
fig = plt.figure(figsize=(8, 3))
ax = fig.add_subplot()
plot_phot(df2013, 'mjd', 'scienceMag', 'scienceMagErr', 'band', ax)
ax.legend()
ax.set_xlabel('MJD')
ax.set_ylabel('mag')
plt.gca().invert_yaxis()
plt.title('Photometry of MAXI1803-298 from DECam using LSST Pipelines', fontsize=14)
```

### 2015 DECam data

```python
fig = plt.figure(figsize=(8, 3))
ax = fig.add_subplot()
plot_phot(df2015, 'mjd', 'scienceMag', 'scienceMagErr', 'band', ax)
ax.legend()
ax.set_xlabel('MJD')
ax.set_ylabel('mag')
plt.gca().invert_yaxis()
plt.title('Photometry of MAXI1803-298 from DECam using LSST Pipelines', fontsize=14)
```
